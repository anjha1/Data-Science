âœ… 1. Agglomerative Clustering on Small Dataset

```python
import numpy as np
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
import matplotlib.pyplot as plt
```

```python
# Small sample data
X = np.array([
    [0.678735154857774, 0.655477890177557],
    [0.757740130578333, 0.171186687811562],
    [0.743132468124916, 0.706046088019609],
    [0.392227019534168, 0.0318328463774207]
])
```

```python
# Create linkage using single linkage and chebyshev distance
Z = linkage(X, method='single', metric='chebyshev')
```

```python
# Form a maximum of 3 clusters
T = fcluster(Z, t=3, criterion='maxclust')
```

```python
# Plot dendrogram with appropriate cutoff
cutoff = np.median([Z[-3, 2], Z[-2, 2]])
plt.figure(figsize=(10, 5))
dendrogram(Z, color_threshold=cutoff)
plt.title('Dendrogram (Single Linkage, Chebyshev Distance)')
plt.show()
```

âœ… 2. Agglomerative Clustering on Large Random Dataset

```python
from mpl_toolkits.mplot3d import Axes3D
from sklearn.datasets import make_blobs

np.random.seed(0)
X_large = np.random.rand(20000, 3)
```

```python

# Create linkage using ward method
Z_large = linkage(X_large, method='ward')
```

```python
# Form 4 clusters
clusters_large = fcluster(Z_large, t=4, criterion='maxclust')
```

```python
# Plot clusters
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_large[:, 0], X_large[:, 1], X_large[:, 2], c=clusters_large, cmap='viridis', s=1)
ax.set_title('3D Scatter Plot of 4 Clusters (Ward Linkage)')
plt.show()
```
