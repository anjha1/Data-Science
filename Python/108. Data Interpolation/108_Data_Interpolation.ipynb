**Data Interpolation**

Data interpolation is the process of estimating unknown values within a dataset based on the known values. In Python, there are various libraries available that can be used for data interpolation, such as NumPy, SciPy, and Pandas. Here is an example of how to perform data interpolation using the NumPy library:



**1.Liner Interpolation**

```python
import numpy as np
x = np.array([1, 2,3, 4, 5])
y = np.array([2,4,6,8,10])
```

```python
x[0]
```

```python
x[-1]
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

**Interpolat the data using Linear Interpolation**

```python
x_new = np.linspace(x[0],x[-1],10)
y_interp = np.interp(x_new,x,y)
x_new
```

```python
y_interp
```

```python
plt.scatter(x_new,y_interp)
```

```python
plt.scatter(x,y)
plt.plot(x_new,y_interp,color='red')
```

**Cubic Interpolation with SciPy**

```python
x = np.array([1, 2,3, 4, 5])
y = np.array([3,8,27,64,125])
```

```python
from scipy.interpolate import interp1d
f = interp1d(x,y,kind='cubic')
x_new = np.linspace(x[0],x[-1],10)
y_interp = f(x_new)
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x_new,y_interp)
```

**Polynomial Interpolation**

```python
x = np.array([1, 2,3, 4, 5])
y = np.array([1,4,9,16,25])

#Interpoleate the data using Polynomial Interpolation

p=np.polyfit(x,y,2) #fit a 2nd Degree polynomial to the data

x_new = np.linspace(x[0],x[-1],10)
y_interp = np.polyval(p,x_new)
```

```python
plt.scatter(x,y)
```

```python

plt.scatter(x_new,y_interp,color='red')
```

```python
plt.scatter(x,y)
plt.scatter(x_new,y_interp,color='red')
```

```python
x1 = np.array([1, 2,3, 4, 5])
y1 = np.array([1,32,27,16,25])

#Interpoleate the data using Polynomial Interpolation

p1=np.polyfit(x1,y1,5) #fit a 5nd Degree polynomial to the data

x_new1 = np.linspace(x1[0],x1[-1],30)
y_interp1 = np.polyval(p1,x_new1)
```

```python
plt.scatter(x1,y1)
plt.scatter(x_new1,y_interp1,color='red')
```
