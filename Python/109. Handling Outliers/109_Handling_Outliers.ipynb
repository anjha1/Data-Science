# **Handling Outliers**

**5 number summary**
1. minimum value
2. Q1 (25) percentile
3. median
4. Q3 (75) percentile
5. Maximum value

```python
import numpy as np
list_marks=[42,32,56,75,89,54,32,89,90,87,67,54,45,98,99,67,74,1000,1100]
np.percentile(list_marks,[25])
```

```python
np.percentile(list_marks,[50])
```

```python
np.percentile(list_marks,[75])
```

```python
np.percentile(list_marks,[100])
```

```python
min,Q1,Q2,Q3,max=np.percentile(list_marks,[0,25,50,75,100])
```

```python
print(min,Q1,Q2,Q3,max)
```

```python
min,Q1,Q2,Q3,max=np.quantile(list_marks,[0.0,0.25,0.50,0.75,1.0])
```

```python
print(min,Q1,Q2,Q3,max)
```

```python
IQR=Q3-Q1
IQR
```

```python
Lower_fence=Q1-1.5*IQR
Upper_fence=Q3+1.5*IQR
```

```python
Lower_fence,Upper_fence
```

```python
outliers=[]
non_outliers=[]
for i in list_marks:
  if i<Lower_fence or i>Upper_fence:
    outliers.append(i)
  else:
    non_outliers.append(i)
```

```python
print(non_outliers)
```

```python
print(outliers)
```
