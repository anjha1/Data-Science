**Imbalanced Dataset Handling**



1.   Upsampling
2.   Down Sampling



```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(123)   # the same sequence of random numbers is generated every time you run this code.
n_samples = 1000
class_0_ratio = 0.9
n_class_0 = int(n_samples * class_0_ratio)
n_class_1 = n_samples - n_class_0
```

```python
n_class_0,n_class_1
```

```python
class_0 = pd.DataFrame({
    'feature_1': np.random.normal(loc=0, scale=1, size=n_class_0),
    'feature_2': np.random.normal(loc=0, scale=1, size=n_class_0),
    'target': [0] * n_class_0
})

class_1 = pd.DataFrame({
    'feature_1': np.random.normal(loc=2, scale=1, size=n_class_1),
    'feature_2': np.random.normal(loc=2, scale=1, size=n_class_1),
    'target': [1] * n_class_1
})

```

```python
class_0
```

```python
class_1
```

```python
df = pd.concat([class_0, class_1]).reset_index(drop=True)
df.head()
```

```python
df.tail()
```

```python
df['target'].value_counts()
```

**Upsampling**

```python
df_minority=df[df['target']==1]
df_majority=df[df['target']==0]
```

```python
df_minority.head()
```

```python
df_majority.head()
```

**Upsampling Perform**

```python
from sklearn.utils import resample
df_minority_upsampled = resample(df_minority,
                                  replace=True, # sample with replacement
                                  n_samples=len(df_majority), # match number in majority class
                                  random_state=42
                                  )
```

```python
df_minority_upsampled.shape
```

```python
df_minority_upsampled.head()
```

```python
df_upsampled = pd.concat([df_majority, df_minority_upsampled])
```

```python
df_upsampled['target'].value_counts()
```

```python
df_upsampled.shape
```

**Down Sampling**

```python
np.random.seed(123)   # the same sequence of random numbers is generated every time you run this code.
n_samples = 1000
class_0_ratio = 0.9
n_class_0 = int(n_samples * class_0_ratio)
n_class_1 = n_samples - n_class_0
```

```python
class_0 = pd.DataFrame({
    'feature_1': np.random.normal(loc=0, scale=1, size=n_class_0),
    'feature_2': np.random.normal(loc=0, scale=1, size=n_class_0),
    'target': [0] * n_class_0
})

class_1 = pd.DataFrame({
    'feature_1': np.random.normal(loc=2, scale=1, size=n_class_1),
    'feature_2': np.random.normal(loc=2, scale=1, size=n_class_1),
    'target': [1] * n_class_1
})

```

```python
df1 = pd.concat([class_0, class_1]).reset_index(drop=True)
df1.head()
```

```python
df1['target'].value_counts()
```

```python
df1_minority=df1[df1['target']==1]
df1_majority=df1[df1['target']==0]
```

```python
from sklearn.utils import resample
df1_majority_upsampled = resample(df1_majority,
                                  replace=False, # sample without replacement
                                  n_samples=len(df1_minority), # match number in minority class
                                  random_state=42
                                  )
```

```python
df1_minority.shape
```

```python
df1_majority.shape
```

```python
df1_majority_upsampled.shape
```

```python
df1_downsampled = pd.concat([df1_minority, df1_majority_upsampled])
```
