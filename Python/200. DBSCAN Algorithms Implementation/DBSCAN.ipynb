# **DBSCAN Algorithms**

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
%matplotlib inline
```

[link make_moons](https://scikit-learn.org/0.16/modules/generated/sklearn.datasets.make_moons.html)

```python
X,y=make_moons(n_samples=250,noise=0.05)
```

```python
X
```

```python
plt.scatter(X[:,0],X[:,1])
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
```

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan=DBSCAN(eps=0.5)
```

[link DBSCAN](https://scikit-learn.org/dev/modules/generated/sklearn.cluster.DBSCAN.html)

```python
dbscan.fit(X_scaled)
```

```python
dbscan.labels_
```

```python
plt.scatter(X_scaled[:,0],X_scaled[:,1],c=dbscan.labels_)
```

```python
plt.scatter(X[:,0],X[:,1],c=dbscan.labels_)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```
