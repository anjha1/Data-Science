```python
from sklearn.datasets import load_iris
```

```python
dataset=load_iris()
```

```python
dataset
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
dataset.keys()
```

```python
df=pd.DataFrame(dataset.data,columns=dataset.feature_names)
```

```python
df.head()
```

```python
df['target']=dataset.target
```

```python
df.head()
```

```python
dataset.target_names
```

```python
dataset.target
```

#Binary Classification

```python
df_copy=df[df['target']!=2]
```

```python
df_copy.head()
```

```python
df_copy['target'].value_counts()
```

# independent and dependent feature

```python
X=df_copy.iloc[:,:-1]
y=df_copy.iloc[:,-1]
```

#train test split

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
classifier=LogisticRegression()
classifier.fit(X_train,y_train)
```

#Prediction

```python
classifier.predict_proba(X_test)
```

```python
y_pred=classifier.predict(X_test)
```

```python
y_pred
```

## Confusion matrix, accuracy score, classification report


```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

#Hyperparameter Tuning


## Gridsearchcv


```python
from sklearn.model_selection import GridSearchCV
import warnings
warnings.filterwarnings('ignore')
```

```python
# Defining parameter grid
parameters = {
    'penalty': ['l1', 'l2', 'elasticnet'],  # Corrected invalid values
    'C': [1, 10, 20]
}

# Creating GridSearchCV instance
clf = GridSearchCV(estimator=classifier, param_grid=parameters, cv=5)

# Splitting training data to validation data
clf.fit(X_train, y_train)
```

```python
clf.best_params_
```

```python
classifier=LogisticRegression(C=1,penalty='l2')
classifier.fit(X_train,y_train)
```

```python
y_pred=classifier.predict(X_test)
```

```python
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
# Importing required libraries
from sklearn.model_selection import RandomizedSearchCV


# Creating RandomizedSearchCV instance
random_clf = RandomizedSearchCV(
    estimator=LogisticRegression(),
    param_distributions=parameters,
    cv=5
)

# Fitting the model
random_clf.fit(X_train, y_train)

# Displaying best parameters
random_clf.best_params_
```

```python
classifier=LogisticRegression(C=1,penalty='l2')
classifier.fit(X_train,y_train)
```

```python
y_pred=classifier.predict(X_test)
```

```python
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
