# **Covariance And Correlation**

```python
import seaborn as sns
```

```python
df=sns.load_dataset('healthexp')
df
```

```python
# Select only numeric columns before calculating covariance
numerical_df = df.select_dtypes(include=['number'])

# Calculate covariance for the numeric columns only
covariance_matrix = numerical_df.cov()

covariance_matrix
```

# **Pearson correlation coefficient**

```python
numerical_df.corr(method='pearson')
```

# **Spearman's rank correlation coefficient**

```python
numerical_df.corr(method='spearman')
```

```python
numerical_df.var()
```
